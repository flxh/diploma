```python
import  numpy as np
import matplotlib.pyplot as plt

all_labels = ['RL', 'Priorität']
all_nameschemes = [
                      '../ml/evaluations/{}.csv',
                      '../simulation/mat_evaluations/{}_prio.csv'
                  ]
n_plot_steps = 21*96

for i in range(10):
    fig, (ax1,ax2,ax3) = plt.subplots(3,1, figsize=(14,14))
    print(i)

    for d, name_scheme in zip(all_labels, all_nameschemes):
        data = np.genfromtxt(name_scheme.format(i), delimiter=';')
        ax1.plot(range(n_plot_steps), (data[:,1]+data[:,2])[:n_plot_steps]/900)
        ax2.plot(range(n_plot_steps), data[:,3][:n_plot_steps] / 900)
        ax3.plot(range(n_plot_steps), data[:,4][:n_plot_steps])
    ax1.legend(['RL', 'Prio'])
    plt.show()
```
