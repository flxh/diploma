```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 8,'lines.markersize':2})


df_all = pd.read_csv('/workspace/total_eval_summary.csv', ';')
fig, ax = plt.subplots(figsize=(6.2,3))
for i in range(0,12,2):
    mask = pd.notna(df_all.iloc[: ,3+i])
    df = df_all.iloc[: ,[0,3+i]][mask]
    df.iloc[:,1] = df.iloc[:,1].rolling(60).mean()
    ax.plot(df.iloc[:,0],df.iloc[:,1],  label=i)
    ax.legend([
        '${\\alpha=1\\times10^{-4}}; \\gamma=0.99$',
        '${\\alpha=5\\times10^{-4}}; \\gamma=0.99$',
        '${\\alpha=1\\times10^{-4}}; \\gamma=0.997$',
        '${\\alpha=5\\times10^{-4}}; \\gamma=0.997$',
        '${\\alpha=1\\times10^{-4}}; \\gamma=0.9995$',
        '${\\alpha=5\\times10^{-4}}; \\gamma=0.9995$',
    ])
    ax.set_ylabel('Evalutation Score')
    ax.set_xlabel('$i_{iter}$')

plt.tight_layout()
#plt.savefig('/home/florus/Documents/Uni Arbeit/Veröffentlichung/images/hyperparameter_study.pdf')
plt.show()
```
