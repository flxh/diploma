```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_all = pd.read_csv('E:/workspace/total_eval_summary.csv', ';')
fig, ax = plt.subplots(figsize=(10,10))
for i in range(0,12):
    mask = pd.notna(df_all.iloc[: ,3+i])
    df = df_all.iloc[: ,[0,3+i]][mask]
    df.iloc[:,1] = df.iloc[:,1].rolling(30).mean()
    ax.plot(df.iloc[:,0],df.iloc[:,1],  label=i)
    ax.legend()
```
